```python

df_filtered = df

#getting first 5 columns
df_filtered = df_filtered.iloc[:, :7]

#df_filtered = df_filtered.drop(columns=['name', 'altCount', "shiftBases", 'freqSourceCount', 'minorAlleleFreq', 'majorAllele',
       #'minorAllele', 'maxFuncImpact', 'class', 'ucscNotes', '_dataOffset',
       #'_dataLen'])
```

```python
df_filtered
```

```python
# removing commas in alts
df_filtered['alts'] = df_filtered['alts'].str.replace(',', '', regex=False)
df_filtered
```

```python
# filtering out rows where ref or alt has length > 1
df_filtered = df_filtered[(df_filtered['ref'].str.len() == 1) & (df_filtered['alts'].str.len() == 1)]
df_filtered
```

```python
# subtracting chromStart and chromEnd by 1
df_filtered['chromStart'] = df_filtered['chromStart'] - 1
df_filtered['chromEnd'] = df_filtered['chromEnd'] - 1
```

```python
# if there are _ in chrom, take the first part (up until the first _)
df_filtered['#chrom'] = df_filtered['#chrom'].str.split('_').str[0]
df_filtered
```

```python
#final df should have chrom, chromstart, chromend, ref, and alts
df_filtered = df_filtered[['#chrom', 'chromStart', 'chromEnd', 'ref', 'alts']]
```

```python
#exporting as bed

df_filtered.to_csv("/home/ubuntu/honors_research/ucsc_snps/ucsc_snps.bed", index=False, header = False, sep='\t')
```
